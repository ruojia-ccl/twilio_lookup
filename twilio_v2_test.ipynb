```python
import pandas as pd
import asyncio
from twilio_lookup_v2 import TwilioLookup
```

```python
df = pd.read_csv(
    "twilio_test_data.csv",
    sep="\t",
    dtype=str,
    usecols=["LIKELYCELLPHONE"],
)

test_phones = (
    df["LIKELYCELLPHONE"]
    .dropna()
    .str.strip()
    .loc[lambda s: s.str.len() == 10]
    .head(50)
    .tolist()
)

print(f"Loaded {len(test_phones)} phones")
print(test_phones[:5])
```

```python
lookup = TwilioLookup()
results = await lookup.validate(test_phones)
# results = await lookup.validate(test_phones, force=True)
```

```python
print("=== Valid ===")
display(results["valid"])

print("=== Invalid ===")
display(results["invalid"])

print("=== Failed ===")
display(results["failed"])
```
